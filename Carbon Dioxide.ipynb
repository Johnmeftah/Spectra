```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from os import name 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math 
```

```python
data = pd.read_csv('https://spectralworkbench.org/spectrums/latest/226921.csv')
```

```python
plt.figure(figsize=(14,9.5), dpi=400)
plt.title("Carbon Dioxide Spectrum") 
plt.xlabel("Wavelength (nm)")  
plt.ylabel("Intensity")
plt.plot(data.Wavelength, data.Average, color="black",label='average',
         linewidth=1)
plt.savefig('Carbon Dioxide Spectrum.jpg', format='jpg', dpi=500)
```

```python
plt.figure(figsize=(14,9.5), dpi=400)
plt.title("Carbon Dioxide, Red Color") 
plt.xlabel("Wavelength (nm)")  
plt.ylabel("Intensity")
plt.plot(data.Wavelength, data.Red, color="Red",label='Red', 
         linewidth=1)
plt.savefig('Carbon Dioxide Spectrum1.jpg', format='jpg', dpi=500)
```

```python
data2 = np.loadtxt("https://spectralworkbench.org/spectrums/latest/226921.csv",delimiter=',',skiprows=1)
print(data)
```

```python
plt.figure(figsize=(14,9.5), dpi=400)
plt.title("Carbon Dioxide, Average Counts") 
plt.xlabel("Wavelength (nm)")  
plt.ylabel("Intensity")
wavelengthData = data2[:,0]
averageCounts = data2[:,1]
plt.plot(wavelengthData,averageCounts,'k',label='data')
plt.savefig('Carbon Dioxide Spectrum2.jpg', format='jpg', dpi=500)
```

```python
plt.figure(figsize=(14,9.5), dpi=400)
plt.title("Carbon Dioxide, Red Counts") 
plt.xlabel("Wavelength (nm)")  
plt.ylabel("Intensity")
wavelengthData = data2[:,0]
RedCounts = data2[:,2]
plt.plot(wavelengthData,RedCounts,'r',label='data')
plt.savefig('Carbon Dioxide Spectrum3.jpg', format='jpg', dpi=500)
```

```python
plt.figure(figsize=(14,9.5), dpi=400)
plt.title("Carbon Dioxide, Green Counts") 
plt.xlabel("Wavelength (nm)")  
plt.ylabel("Intensity")
wavelengthData = data2[:,0]
GreenCounts = data2[:,3]
plt.plot(wavelengthData,GreenCounts,'g',label='data')
plt.savefig('Carbon Dioxide Spectrum4.jpg', format='jpg', dpi=500)
```
